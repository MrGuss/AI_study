```python
import pandas as pd
import random

a = pd.Series([random.random()*10, random.random()*10, random.random()*10])
b = pd.Series([random.random()*10, random.random()*10, random.random()*10])

print(a,b,sep="\n\n")

print("distance is ", ((a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2)**0.5)
```
