```python
import math

print("What function do you want to calculate?")
print("1.x+y")
print("2.x-y")
print("3.x*y")
print("4.x/y")
print("5.exp(x+y)")
print("6.sin(x+y)")
print("7.cos(x+y)")
print("8.x^y")
try:
    choice = int(input())
except:
    print("You entered something wrong")

if not(0<choice<9):
    print("You entered unknown number")
    exit
print("Enter x: ")
try:
    x = float(input())
except:
    print("Not a number entered")
print("Enter y: ")
try:
    y = float(input())
except:
    print("Not a number entered")

match choice:
    case 1:
        print(f"{x} + {y} = {x+y}")
    case 2:
        print(f"{x} - {y} = {x-y}")
    case 3:
        print(f"{x} * {y} = {x*y}")
    case 4:
        print(f"{x} / {y} = {x/y}")
    case 5:
        print(f"exp({x} + {y}) = {math.exp(x+y)}")
    case 6:
        print(f"sin({x} + {y}) = {math.sin(x+y)}")
    case 7:
        print(f"cos({x} + {y}) = {math.cos(x+y)}")
    case 8:
        print(f"{x}^{y} = {x**y}")
```
