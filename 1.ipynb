```python
#ex4.2
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,11)
y = np.abs(np.cos(x*np.exp(np.cos(x) + np.log(x+1))))

plt.grid()
plt.plot(x, y, c="r")
plt.fill_between(x, y)

area = np.trapz(y)
print(area)
```
